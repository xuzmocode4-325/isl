```python
# Use the pd.plotting.scatter_matrix() function to produce a scatterplot matrix of the frst columns [Top10perc, Apps, Enroll].
pd.plotting.scatter_matrix(college[["Top10perc", "Apps", "Enroll"]])
```

```python
# Use the boxplot() method of college to produce side-by-side boxplots of Outstate versus Private
college.boxplot("Outstate", "Private")
```

```python
# Create a new qualitative variable, called Elite
college['Elite'] = pd.cut(
    college['Top10perc'], 
    [0,50,100],
    labels=['No', 'Yes']
)
```

```python
#Use the value_counts() method of college['Elite'] to see how many elite universities there are.
college['Elite'].value_counts() 
```

```python
#Use the plot.hist() method of college to produce some histograms with difering numbers of bins for a few of the quantitative variables
from matplotlib import pyplot as plt
number_cols = college.select_dtypes(include=['int64', 'float64']).columns
fig, axs = plt.subplots(6, 3, figsize=(18, 15))
for ax, col in zip(axs.flat, number_cols): 
    ax.set_title(f"{col} histogram")
    ax.hist(college[col])
fig.tight_layout()
```
