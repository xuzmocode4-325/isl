> A very flexible model has less bias and is able to more acturately predict outputs that have a more non-linear relationship to their inputs. The more flexible a model, the more observations it needs to train. Flexible models are more computationally expensive and time consuming to train. Less flexible models, have more bias, and are not able to acurately capure complex relationships between inputs and outputs. They take fewer observations to train and are easier to interprate than more flexible models. A less flexible method might be preffered for inference. 6. Describe the diferences between a parametric and a non-parametric statistical learning approach. What are the advantages of a parametric approach to regression or classifcation (as opposed to a nonparametric approach)? What are its disadvantages?

> Parametric models compute a function *f* based on an assumed form (e.g linear) to which the predictors are considerd coefficients. Non-parametric models do not assume a form and treat each predictor as a dimension. Parametric approaches are less computationally expensive and work well with datasets that have a small amount of observations and many predictors. Non-parametric approaches work best when there are few predictors, as each predictor adds a new dimension to *f*. Non-parametric models also need many more observations than parametric models to make an accurate estimate of the functional form between input variables and their output.
